```python
import os

def replace_occurrences(directory, search_text, replace_text, encoding='latin1'):
    for root, dirs, files in os.walk(directory):
        for file in files:
            file_path = os.path.join(root, file)
            with open(file_path, 'r', encoding=encoding) as f:
                lines = f.readlines()
            with open(file_path, 'w', encoding=encoding) as f:
                for line in lines:
                    f.write(line.replace(search_text, replace_text))
```

```python
def clear_log_file(file_path):
    try:
        # Open the file in write mode ('w') to truncate its contents
        with open(file_path, 'w') as file:
            file.truncate(0)  # Truncate the file to zero length
        print(f"Log file '{file_path}' cleared successfully.")
    except IOError as e:
        print(f"Error: Unable to clear log file '{file_path}'. {e}")
```
